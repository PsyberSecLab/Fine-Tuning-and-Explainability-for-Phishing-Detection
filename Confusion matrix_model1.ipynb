```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Read the data
df = pd.read_csv('Model_experiment.csv')

# True Positives  
tp = df[(df['ground truth'] == "Phishing") & (df['choice made/decision'] == "Phishing")].shape[0]

# True Negatives  
tn = df[(df['ground truth'] == "Ham") & (df['choice made/decision'] == "Ham")].shape[0]   

# False Positives   
fp = df[(df['ground truth'] == "Ham") & (df['choice made/decision'] == "Phishing")].shape[0]

# False Negatives
fn = df[(df['ground truth'] == "Phishing") & (df['choice made/decision'] == "Ham")].shape[0]

# Total
total = tp + tn + fp + fn

# Calculate percentages
tp_percent = (tp / total)
tn_percent = (tn / total)
fp_percent = (fp / total)
fn_percent = (fn / total)

# Create confusion matrix  
matrix = [[tp_percent, fn_percent], [fp_percent, tn_percent]]

# Create DataFrame for confusion matrix
df_cm = pd.DataFrame(matrix, index=['Phishing (ground truth)', 'Ham (ground truth)'], columns=['Predicted Phishing', 'Predicted Ham'])

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(df_cm, annot=True, fmt=".2f", cmap="YlGnBu")

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix')

# Calculate overall accuracy
overall_accuracy = (tp + tn) / (tp + tn + fp + fn) * 100

# Display accuracy
plt.text(0.5, 1.1, f'Accuracy: {overall_accuracy:.2f}%', horizontalalignment='center', verticalalignment='center', transform=plt.gca().transAxes)

plt.savefig("Accuracy_model_choice_over_email_type_model_1.jpg")    
plt.show()
```
